### Plotting with keyword strings

```python
data = {
    'a': np.arange(50),
    'c': rg.integers(0, 50, 50),
    'd': rg.standard_normal(50)
}
data['b'] = data['a'] + 10 * rg.standard_normal(50)
data['d'] = np.abs(data['d']) * 100
plt.scatter('a', 'b', c='c', s='d', data=data) # c: color, s: size
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```

### Plotting with categorical variables

```python
names = ["group_a", "group_b", "group_c"]
values = [1, 10, 100]
plt.figure(figsize=(9, 3))
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle("Categorical Plotting")
plt.show()
```

### Controlling line properties

```python
x = np.arange(10)
y = np.sqrt(np.arange(10))
plt.plot(x, y, linewidth=2.0)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.subplot(121)
line, = plt.plot(x, y, '-')
line.set_antialiased(True)
plt.subplot(122)
line, = plt.plot(x, y, '-')
line.set_antialiased(False)
plt.show()
```

`plt.<method>(*args, **kwargs)`에서 args 또는 kwargs에다가 넣는 사항을 `Line2D` object의 propertiy를 통해 수정할 수도 있다.

```python
x = np.arange(30).reshape(3, 10)
y = x ** 2
lines = [plt.plot(x[i], y[i]) for i in range(3)]
plt.setp(lines[0], color='r', linewidth=1.0)
plt.setp(lines[1], 'color', 'g', 'linewidth', 1.0) # MATLAB style string value pairs
plt.setp(lines[2], color='b', linewidth=1.0)
plt.show()
```

### Working with multiple figures and axes

```python
def f(t): return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.10)
t2 = np.arange(0.0, 5.0, 0.02)
plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

`subplot(numrows, numcols, plot_number)`

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1, 2, 3])
plt.subplot(212)
plt.plot([4, 5, 6])

plt.figure(2)
plt.plot([7, 8, 9])

plt.figure(1)
plt.subplot(211)
plt.title("Easy as 1, 2, 3")
plt.show()
```

### Working with text

`axes([left, bottom, width, height])`

```python
mu, sigma = 100, 15
x = mu + sigma * rg.standard_normal(10_000)
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)

plt.ylabel("Probability")
text = plt.text(60, 0.025, r"$\mu=100,\ \sigma=15$")
plt.setp(text, size=10)
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

### Logarithmic and other nonlinear axes
`matplotlib.scale`: https://matplotlib.org/stable/api/scale_api.html#module-matplotlib.scale

```python
y = rg.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure()
# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title("linear")
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title("log")
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title("symlog")
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title("logit")
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

## matplotlib.pyplot.plot

`fmt`: A format string consists of a part for `marker`, `line` and `color`

Markers: `'.'`, `','`, `'o'`, `'v'`, `'^'`, `'<'`, `'>'`, `'1'`, `'2'`, `'3'`, `'4'`, `'8'`, `'s'`, `'p'`, `'P'`, `'*'`, `'h'`, `'H'`, `'+'`, `'x'`, `'X'`, `'D'`, `'d'`, `'|'`, `'_'`  

Line Styles: `'-'`, `'--'`, `'-.'`, `':'`  

Colors: `'CN'`, `'b'`, `'g'`, `'r'`, `'c'`, `'m'`, `'y'`, `'k'`, `'w'`  
e.g. `'green'`, `'#000000'`

```python
A, B, S, O = 1.5, 1.5, 10, 50
W, H = 3, 3
plt.figure(figsize=(12, 12))
for i in range(1, W * H + 1):
    plt.subplot(W, H, i)
    m = np.array(list('.,ov^<>12348spP*hH+xXDd|_'))
    rg.shuffle(m)
    n = m.size
    c = np.zeros(n, dtype=int)
    for i in range(3):
        b = rg.beta(A, B)
        a = 1 - b
        c += np.floor(rg.beta(a * S, b * S, n) * (1 << 8)).astype(int) * (1 << 8)**i
    s = rg.beta(S, 1, n) + O
    x = np.arange(n)
    y = x.copy()
    rg.shuffle(y)
    for i in range(n):
        plt.scatter(x[i], y[i], c=f"#{hex(c[i])[2:]:0>6}", s=s[i], marker=m[i])
plt.show()
```

```python
from scipy.special import beta, gamma
def plot_beta(a, b):
    x = np.linspace(0.01, 0.99, 99)
    y = (1 / beta(a, b)) * x ** (a - 1) * (1 - x) ** (b - 1)
    plt.plot(x, y)
    plt.show()
plot_beta(1, 4)
```
